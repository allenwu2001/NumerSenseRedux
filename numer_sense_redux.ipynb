## Install Conda and Packages

```python
%%shell
/**************************************/
            INSTALL CONDA
/**************************************/
wget https://repo.anaconda.com/miniconda/Miniconda3-py37_4.12.0-Linux-x86_64.sh
chmod +x Miniconda3-py37_4.12.0-Linux-x86_64.sh
bash ./Miniconda3-py37_4.12.0-Linux-x86_64.sh -b -f -p /usr/local/

conda install --channel defaults conda python=3.7 -y
conda update -n base -c defaults conda -y
conda create -n numersense python=3.7 -y


/**************************************/
        CONDA START ENVIRONMENT 
/**************************************/
source ~/.bashrc 
conda activate numersense

/**************************************/
            CLONE INTO REPO 
/**************************************/
git clone https://github.com/baroodya/NumerSenseRedux.git
cd NumerSenseRedux
mkdir pred_results


/**************************************/
          INSTALLATIONS NEEDED 
/**************************************/
conda install pytorch==1.6.0 cudatoolkit=10.1 -c pytorch -y

pip install transformers==4.28.1
pip install --editable happy-transformer
pip install tensorboardX
pip install sentencepiece
```

# Fine-Tuning

## Bert Large

```python
%%shell
mkdir NumerSenseRedux/saved_models
CUDA_VISIBLE_DEVICES=0 python NumerSenseRedux/src/finetune_mlm.py \
  --output_dir=NumerSenseRedux/saved_models/finetuned_bert_large --overwrite_output_dir \
  --model_type=bert \
  --model_name_or_path=bert-large-uncased \
  --do_train \
  --train_data_file=NumerSenseRedux/data/train.truth.txt  \
  --do_eval \
  --eval_data_file=NumerSenseRedux/data/wiki_complete.txt \
  --per_gpu_train_batch_size 64 \
  --per_gpu_eval_batch_size 64 \
  --block_size 64 \
  --logging_steps 100 \
  --num_train_epochs 3 \
  --line_by_line --mlm 
```

```python
%%shell
python NumerSenseRedux/src/mlm_predict.py \
        reload_bert:NumerSenseRedux/saved_models/finetuned_bert_large \
        NumerSenseRedux/data/validation.masked.txt \
        NumerSenseRedux/pred_results/ft-bert-large.validation.output.jsonl
```

```python
%%shell
python NumerSenseRedux/src/evaluator.py \
        NumerSenseRedux/pred_results/ft-bert-large.validation.output.jsonl \
        NumerSenseRedux/data/validation.masked.tsv
```
